**Dataset:**

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a sample dataset
np.random.seed(0)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
products = ['Product A', 'Product B', 'Product C']
prices = {'Product A': 10, 'Product B': 20, 'Product C': 30}

data = {
    'Date': [random.choice(dates) for _ in range(1000)],
    'Product': [random.choice(products) for _ in range(1000)],
    'Price': [prices[random.choice(products)] for _ in range(1000)],
    'Quantity': [random.randint(1, 10) for _ in range(1000)]
}
```

```python
# Creating DataFrame
sales_df = pd.DataFrame(data)

# Calculating Total sales
sales_df['Total'] = sales_df['Price'] * sales_df['Quantity']

# Saving dataset to CSV
sales_df.to_csv('sales_data.csv', index=False)

# Displaying first few rows of the dataset
print(sales_df.head())
```

**Data Visualizations:**

```python
# Histogram of Total sales
plt.figure(figsize=(6, 4))
sns.histplot(sales_df['Total'], bins=20, kde=True)
plt.title('Distribution of Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram of Price
plt.figure(figsize=(6, 4))
sns.histplot(sales_df['Price'], bins=10, kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Line plot of total sales over time
plt.figure(figsize=(8, 4))
sales_df.groupby('Date')['Total'].sum().plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Bar plot of total sales by product
plt.figure(figsize=(6, 4))
sns.barplot(x='Product', y='Total', data=sales_df, estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Box plot of Quantity
plt.figure(figsize=(6, 4))
sns.boxplot(x='Product', y='Quantity', data=sales_df)
plt.title('Box plot of Quantity by Product')
plt.xlabel('Product')
plt.ylabel('Quantity')
plt.show()
```

```python
# Scatter plot of Quantity vs Total sales
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Quantity', y='Total', data=sales_df)
plt.title('Scatter plot of Quantity vs Total Sales')
plt.xlabel('Quantity')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Pairplot of numerical variables
sns.pairplot(sales_df[['Price', 'Quantity', 'Total']])
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()
```
